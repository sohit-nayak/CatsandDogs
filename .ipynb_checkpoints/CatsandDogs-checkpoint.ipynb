```python
import numpy as np
import os
import matplotlib.pyplot as plt
import cv2
import random
import pickle
```

```python
curr_dir = os.path.join(os.getcwd(), 'PetImages')
categories = ['Dog', 'Cat']
for category in categories:
    path = os.path.join(curr_dir, category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap='gray')
        break
```

```python
IMG_SIZE = 100
new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(new_array, cmap='gray')
```

```python
training_data = []
IMG_SIZE = 50

def create_training_set():
    curr_dir = os.path.join(os.getcwd(), 'PetImages')
    categories = ['Dog', 'Cat']
    for category in categories:
        path = os.path.join(curr_dir, category)
        num_cat = categories.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                training_data.append([new_array, num_cat])
            except Exception as e:
                pass
            
create_training_set()
```

```python
len(training_data)
```

```python
random.shuffle(training_data)
```

```python
X = []
y = []

for features, label in training_data:
    X.append(features)
    y.append(label)
    
```

```python
X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
pickle_out = open('X.pickle', 'wb')
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open('y.pickle', 'wb')
pickle.dump(y, pickle_out)
pickle_out.close()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, MaxPooling2D, Conv2D, Flatten, Dropout

# X = pickle.load(open('X.pickle', 'rb'))
# y = pickle.load(open('y.pickle', 'rb'))
```

```python
X = X / 255.0
X = np.array(X)
y = np.array(y)

model = Sequential()
model.add(Conv2D(64, (3, 3), input_shape=X.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3, 3), input_shape=X.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3, 3), input_shape=X.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])

model.fit(X, y, batch_size=32, validation_split=0.1, epochs=10)
```
